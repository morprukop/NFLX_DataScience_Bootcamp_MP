```python
minprop = pd.read_csv("property_sales.csv")
minprop.info()
```

```python
Adjusted_S = minprop["Adjusted_S"]
px.histogram(Adjusted_S)
```

```python
print(sts.normaltest(Adjusted_S))
# since the p-value is lower than 0.5, we reject the our hypothesis that the adjusted_S of the property sales is normal
```

```python
sample = np.random.choice(Adjusted_S, 5000)
px.histogram(sample)
```

```python
sample = np.random.choice(Adjusted_S, 5000)
print(sts.normaltest(sample))
# The random sample of the Adjusted S is not normal 
```
